```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_animes['tags'] = new_animes['tags'].apply(stem)
```

```python
new_animes['tags'] = new_animes['tags'].apply(lambda x:x.lower())
new_animes['tags'] = new_animes['tags'].apply(lambda x:x.replace("[written by mal rewrite]", ""))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=8000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_animes['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[481])), reverse=True, key= lambda x:x[1])[1:11]
```

```python
def recommendCS(anime):
    if not new_animes[new_animes['title'] == anime].empty:
        anime_index = new_animes[new_animes['title'] == anime].index[0]
        distances = similarity[anime_index]
        animes_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:31]

        for i in animes_list:
            print(new_animes.iloc[i[0]].title, new_animes.iloc[i[0]].mal_id)
    else:
        print("Chưa có anime/manga này trong dữ liệu")
```

```python
recommendCS('SKET Dance')
```

```python
# Thuat toan KNN
vectorizer = TfidfVectorizer(max_features=8000,stop_words='english')
features = vectorizer.fit_transform(new_animes['tags'])
```

```python
# Create kNN model (experiment with different k values)
knn = NearestNeighbors(n_neighbors=31, metric='euclidean')
knn.fit(features)
```

```python
def recommendKNN(anime):
    if not new_animes[new_animes['title'] == anime].empty:
        anime_index = new_animes[new_animes['title'] == anime].index[0]
        print(anime_index)
        anime_tags = new_animes[new_animes['title'] == anime].tags[anime_index]
        anime_tags_vector = vectorizer.transform([anime_tags])[0]
        distances, indices = knn.kneighbors(anime_tags_vector.reshape(1, -1))
        animes_list = new_animes.loc[indices[0][1:]]

        # for i in animes_list:
        #     print(new_animes.iloc[i[0]].title, new_animes.iloc[i[0]].mal_id)
        for i, row in animes_list.iterrows():
            print(f"- {row['title']}")
    else:
        print("Chưa có anime/manga này trong dữ liệu")
```

```python
recommendKNN('SKET Dance')
```
